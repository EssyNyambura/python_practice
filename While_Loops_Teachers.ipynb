# 🌀 Python Lesson: While Loops

In this lesson, you'll learn how to use `while` loops in Python — a key tool for repeating actions **until** a certain condition is met.



## 🧩 What is a `while` loop?
A `while` loop keeps running **as long as** a certain condition is true.

**Syntax:**
```python
while condition:
    # code block
```
When the condition becomes `False`, the loop stops.


````
# Example: Counting from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
print("Done!")
````


## ⚠️ Common Mistake: Infinite Loops
Always make sure something inside the loop eventually makes the condition `False`!


````
# ❌ Infinite loop example (do not run!)
# count = 0
# while count < 5:
#     print("Still counting...")
#     # forgot to increment count
````


## 🔄 Breaking Out of a Loop Early
Use `break` to stop a loop even if the condition is still true.


````
num = 1
while num <= 10:
    if num == 6:
        break
    print(num)
    num += 1
print("Stopped at", num)
````


## 🔁 Skipping an Iteration
Use `continue` to skip the rest of the loop and move to the next iteration.


````
x = 0
while x < 10:
    x += 1
    if x % 2 == 0:
        continue
    print(x)
````

## ✅ Exercise Solutions


### Solution 1 — Countdown

````
i = 10
while i > 0:
    print(i)
    i -= 1
print("Blast off!")
````

### Solution 2 — User Input Loop

````
num = int(input("Enter a number: "))
while num != 0:
    num = int(input("Enter a number: "))
print("Goodbye!")
````

### Solution 3 — Summing Numbers

````
total = 0
while True:
    user_input = input("Enter a number (or 'stop'): ")
    if user_input.lower() == "stop":
        break
    total += int(user_input)
print("Total sum =", total)
````

### Solution 4 — Guessing Game 🎯

````
import random
secret = random.randint(1, 10)
guess = 0

while guess != secret:
    guess = int(input("Guess the number: "))
    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print("Correct! 🎉")
````
