```python
import numpy as np
print(np.random.binomial(10, 0.5, size = 10))
print(np.random.hypergeometric(10, 20, 5, size = 10))
print(np.random.poisson(5, size = 10))
```

```python
#funtion to reverse a string
def reverse_string(s):
    i = len(s) - 1
    result=''
    while i >= 0:
        result += s[i]
        i -= 1
    return result
print(reverse_string('Python'))
```

```python
#function to return the inner elements of a list
def inner_list(lst):
    return lst[1:-1]
print(inner_list([1, 2, 3, 4, 5]))
print(inner_list(['a', 'b', 'c', 'd', 'e']))
```

```python
#a function to return the sum of even numbers in a list
def sum_of_even_numbers(nums):
    result = 0
    for n in nums:
        if n % 2 == 0:
            result += n
    return result
print(sum_of_even_numbers([1, 2, 3, 4, 5, 6]))
```
