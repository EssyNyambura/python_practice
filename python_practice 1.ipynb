```python
def avg(list): #creates a function avg that takes in the argument list and calculates the average of a list of numbers
    result = 0 # stores the final average
    sum = 0 #stores the sum of all the numbers
    number_of_elements = 0 #tracks how many numbers are in the list
    for l in list: #loop that goes through every number in the list
        number_of_elements += 1 #increase count for each number
        sum += l #adds current number to the running total
        
    result = sum / number_of_elements #gets the average
    return result

print(avg([1,2,3,4,5]))

def stdev(list):
    result = 0
    average = avg(list)
    sum_of_squares = 0
    number_of_elements = 0
    for l in list:
        sum_of_squares += (l- average)**2
        number_of_elements += 1
    result = sum_of_squares/ number_of_elements #variance
    result= result **0.5 #standard deviation
    return result

l1 = [1,2,3,4,5]
print(avg(l1),stdev(l1))
```
